```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
```

## Preprocessing

```python
number_of_words = 20000
max_len = 100
```

```python
(x_t, y_t), (x_T, y_T) = imdb.load_data(num_words=number_of_words)
```

### Pad Data

```python
x_t = tf.keras.preprocessing.sequence.pad_sequences(x_t, maxlen=max_len)
```

```python
x_T = tf.keras.preprocessing.sequence.pad_sequences(x_T, maxlen=max_len)
```

## Building Model

```python
model = tf.keras.Sequential()
```

### Embedding Layer

```python
model.add(tf.keras.layers.Embedding(input_dim=number_of_words, output_dim=128, input_shape=(x_t.shape[1],)))
```

### LSTM Layer

```python
model.add(tf.keras.layers.LSTM(units=128, activation="tanh")) #normally use hyperbolic tangent
```

### Output Layer

```python
model.add(tf.keras.layers.Dense(units=1, activation="sigmoid")) #predicting probability of 1 or possitive, so we use sigmoid
```

## Compile Model

```python
model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["accuracy"]) #rmdprop generally better for RNN
```

```python
model.summary()
```

```python
model.fit(x_t, y_t, epochs=3, batch_size=128)
```

```python
test_loss, test_acc = model.evaluate(x_T, y_T)
```
